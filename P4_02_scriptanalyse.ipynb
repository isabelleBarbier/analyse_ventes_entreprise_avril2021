```python
def filter_month_category(num_mois, num_categorie, df_data):
    filter_mois = df_data['mois'] == int(num_mois)
    filter_categ = df_data['categ'] == int(num_categorie)
    return df_data[filter_mois & filter_categ]
```

```python
def courbeLorentz(array, xlabel, ylabel, labelCourbeLorentz, titre):    
    dep = array.values
    lorenz = np.cumsum(np.sort(dep)) / dep.sum()
    lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
    
    fig, ax = plt.subplots(figsize=[6,6])
# Courbe de Lorentz
    ax = plt.plot(np.linspace(0,1,len(lorenz)),lorenz,drawstyle='steps-post', label=labelCourbeLorentz)
# Ligne d'égalité   
    ax = plt.plot([0,1], [0,1],  linestyle='dashed', color= 'r', label='droite d\'équirépartition')
    plt.title(titre, size=16)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    return plt.show()
```

```python
def indiceGini(array):    
    dep = array.values
    lorenz = np.cumsum(np.sort(dep)) / dep.sum()
    lorenz = np.append([0],lorenz)
    
    aire_ss_courbe = lorenz[:-1].sum()/len(lorenz) # aire sous la courbe de Lorenz. La dernière valeur ne participe pas à l'aire, d'où "[:-1]"
    S = 0.5 - aire_ss_courbe # aire entre la 1e bissectrice et la courbe de Lorenz
    gini = 2*S
    return gini
```

```python
def eta_squared(df,X,Y):
    sous_echantillon = df.copy()
    x = sous_echantillon[X]
    y = sous_echantillon[Y]
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
```

La courbe correspond à une courbe "poisson"

```python
# Loi de POISSON : loi discrète et infinie => peut prendre toutes les valeurs || k=évènement(temps de session) x:var
# Prob que la variable prenne la valeur définie (temps de session)=>P(x=k)
# Prob calcul : parametre = espérance
df_tpsSessionPoisson = df_resNotNull.copy()
df_tpsSessionPoisson.head(5)
```

```python
# Loi de POISSON 
# Espérance
mu = df_tpsSessionPoisson['temps de session (minute)'].mean()
mu
```

```python
# Loi de POISSON 
r = poisson.mean(mu, loc=0)
r
```

```python
# Loi de POISSON 
#  cdf proba cumulative donc proba < à la valeur || pmf : proba= à la valeur
df_tpsSessionPoisson['proba'] = poisson.pmf(df_tpsSessionPoisson['temps de session (minute)'], mu)
```

```python
df_tpsSessionPoisson['proba'] = (df_tpsSessionPoisson['proba']*100).sort_values(ascending=False)
```

- Méthode avec tableau de contingence coloré et chi2_contingency

```python
# Analyse entre   variables qualitatives => chi-2 - tableau de contingence
X = "sex"
Y = "categ"

c = df_sexCat[[X,Y]].pivot_table(index=X,columns=Y,aggfunc=len)
cont = c.copy()

tx = df_sexCat[X].value_counts()
ty = df_sexCat[Y].value_counts()

cont.loc[:,"Total"] = tx
cont.loc["total",:] = ty
cont.loc["total","Total"] = len(df_sexCat)
cont
```

```python
c
```

```python
# Affichage du tableau de contingence coloré
plt.figure(figsize=(12,5))
tx = pd.DataFrame(tx)
ty = pd.DataFrame(ty)
tx.columns = ["foo"]
ty.columns = ["foo"]
n = len(df_sexCat)
indep = tx.dot(ty.T) / n

c = c.fillna(0) # on remplace les valeurs nulles par des 0
mesure = (c-indep)**2/indep
xi_n = mesure.sum().sum()
sns.heatmap(mesure/xi_n,annot=c)
plt.title('Tableau de contingence (genre des clients / catégorie de produits)', size= 16)
plt.xlabel('catégorie des produits achetés', size= 14)
plt.ylabel('genre des clients', size=14)
plt.show()
# plt.savefig('Tableau de contingence (genre des clients / catégorie de produits) An.jpg')
```

```python
X = "age_bins"
Y = "categ"
cont = df_ageCat[[X,Y]].pivot_table(index=X,columns=Y,aggfunc=len,margins=True,margins_name="Total")
cont
```

```python
tx = cont.loc[:,["Total"]]
ty = cont.loc[["Total"],:]
n = len(df_ageCat)
indep = tx.dot(ty) / n

c = cont.fillna(0) # On remplace les valeurs nulles par 0
measure = (c-indep)**2/indep
xi_n = measure.sum().sum()
table = measure/xi_n
plt.figure(figsize=(10,5))
sns.heatmap(table.iloc[:-1,:-1],annot=c.iloc[:-1,:-1])
plt.title('Tableau de contingence (âge des clients / catégorie de produits)', size= 16)
plt.xlabel('catégorie des produits achetés', size= 14)
plt.ylabel('âge des clients', size=14)
# plt.savefig('Tableau de contingence (âge des clients / catégorie de produits)An.jpg')
plt.show()
```
